```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

with open('../I001_Results/DATA_PICK_504_A.pkl', 'rb') as f:
    data = pickle.load(f)

t = np.array(data['t'])
energies = ['ALLIE', 'ALLKE', 'ALLSD', 'ALLSE', 'ALLWK']

fig, ax = plt.subplots(figsize=(9, 5))
for key in energies:
    ax.plot(t, data[key]['ASSEMBLY'], label=key, linewidth=1.8)

ax.set_xlabel('Time [s]')
ax.set_ylabel('Energy [mJ]')
ax.set_title('SIM 504 — All energies')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
